### Financial Anti-Fraud Common Features 



#### User Basic Attributes

phone_number
- Whether the phone number prefixes are the same
- Whether the phone number locations are the same
- Whether it is a virtual operator
- Whether it is a data card or a calling card

nickname
- Whether the nickname follows a fixed pattern (e.g., Chinese characters + numbers)
- Whether the remark matches certain intimate terms

birthday
- Age
- Zodiac sign
- Chinese zodiac

sex
- Whether there is a gender imbalance

password
- Whether passwords are all the same

ID number
- Verify age
- Verify gender ratio
- Verify city

email
- Whether it is a disposable email
- Whether the username follows certain rules
- Whether they use the same email provider
- Data within the email (e.g., bills)

education
- Similarity in educational background

housing
- Whether rental status is identical

points
- Whether the points exceed a certain threshold

check-in
- Similarity in check-in behavior

IP
- Whether the IP addresses belong to the same segment
- Whether each login uses the same IP address
- Whether it is a temporary IP and matches GPS
- Whether IP and GPS match

GPS
- Latitude and longitude similarity analysis
- Similarity in country, province, and city
- Whether IP and GPS match

WiFi
- SSID
- WiFi list
- Whether the WiFi switched a few minutes before applying for a loan

application time
- Time slicing
- How long the registration process took (too fast or too slow can be problematic)
- Total number of applications

login time
- Time slicing
- Number and frequency of logins
- Interval between the last login and loan application time
- Perform checks for simultaneous logins (e.g., multiple users logging in at the same time)

ua (user agent)
- Whether the same user agent is used for each session

channel
- App/H5/WeChat
- Whether the channel ID belongs to a prohibited channel

app version
- Whether the app version is the same each time
- Whether the app version is too outdated (older versions might have bugs that could be exploited by malicious intermediaries)

referrer/contact person
- Name matching
- Phone number matching

#### Device Fingerprint

IMEI
- Whether they are all the same
- Whether the IMEI number is the same for each login

device ID
- Whether they are all the same
- Whether the device ID is the same for each login

screen resolution
- Whether the phone model matches the screen resolution

mobile type
- Phone brand
- Phone model

OS (Operating System)
- Whether the operating system is the same each time it is opened
- Whether the OS is the same for all applicants
- Whether the OS version is too outdated

#### Chinese Typo Matching via Pinyin Similarity

address
- Address should be standardized
- Fuzzy matching
- Similarity calculation (cosine distance, word vectors)

company
- Regular expressions
- Byte splitting
- Keyword extraction
- Similarity calculation
- Typo/homophone recognition
